📝 Redacción del Problema
Se desea desarrollar un programa interactivo en Python, dirigido a estudiantes de nivel preparatoria, que permita realizar cálculos geométricos fundamentales de manera práctica y visual. La aplicación presentará un menú principal con opciones para calcular el área y el perímetro de un círculo, así como el volumen de una esfera. Cada vez que el usuario seleccione una operación, el sistema solicitará el valor del radio correspondiente, ejecutará el cálculo matemático y mostrará el resultado de inmediato. El programa se mantendrá en ejecución de forma cíclica hasta que el usuario decida salir, con el objetivo de reforzar el uso de funciones, estructuras de control (if, elif, else, while) y llamadas a subrutinas en un entorno amigable, sin validaciones complejas, para enfocarse en la lógica básica de programación.

1. 🔷 Área del círculo:
A = π · r²
2. 🔷 Perímetro (circunferencia) del círculo:
P = 2 · π · r
3. 🔷 Volumen de la esfera:
V = (4/3) · π · r³

![texto del vínculo](https://raw.githubusercontent.com/alexandroaguilar128/Mi-imagenes-de-EPPL-Unidad-2-/refs/heads/main/circulo.PNG)



```python
# Elaborado por ✅Alexandro Aguilar
import math

def areacirculo():
    """Calcula y muestra el área de un círculo."""
    radio = float(input("📏 Ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"🟢 El área del círculo es: {area:.2f} unidades²\n")

def perimetrocirculo():
    """Calcula y muestra el perímetro de un círculo."""
    radio = float(input("📏 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"🔵 El perímetro del círculo es: {perimetro:.2f} unidades\n")

def volumenesfera():
    """Calcula y muestra el volumen de una esfera."""
    radio = float(input("📏 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"🟣 El volumen de la esfera es: {volumen:.2f} unidades³\n")

def menu():
    """Muestra el menú interactivo y devuelve la opción elegida."""
    print("\n📐✨ MENÚ GEOMÉTRICO INTERACTIVO ✨📐")
    print("1️⃣  Calcular área del círculo 🟢")
    print("2️⃣  Calcular perímetro del círculo 🔵")
    print("3️⃣  Calcular volumen de la esfera 🟣")
    print("4️⃣  Salir del programa 🚪")
    opcion = input("👉 Selecciona una opción (1-4): ")
    return opcion

# 🔹 PROGRAMA PRINCIPAL
print("🌟 ¡Bienvenido a la Calculadora Geométrica! 🌟")
while True:
    opcion = menu()

    if opcion == "1":
        areacirculo()
    elif opcion == "2":
        perimetrocirculo()
    elif opcion == "3":
        volumenesfera()
    elif opcion == "4":
        print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🚀")
        break
    else:
        print("\n⚠️ Opción no válida. Por favor, elige un número del 1 al 4.\n")
```
